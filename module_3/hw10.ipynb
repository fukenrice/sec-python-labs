```python
import pandas as pd

with open("events.json", "r", encoding="utf-8") as file:
    data = pd.read_json(file)

df = pd.json_normalize(data["events"])

df.head()
```

```python
signature_counts = df["signature"].value_counts()

print(signature_counts.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns.barplot(x=signature_counts.index, y=signature_counts.values, palette="viridis")

plt.xticks(rotation=90)  # Поворачиваем подписи для удобства
plt.xlabel("signature")
plt.ylabel("Count")

plt.show()
```
